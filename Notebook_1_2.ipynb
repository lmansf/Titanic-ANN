```python
df_dropper = df_explore[df_explore['Fare'] >= 0].copy()
df_explore = df_dropper.dropna()
```

```python
X = df_explore.drop(columns=['Survived'])
y = df_explore['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python

scaler = StandardScaler()
# Don't cheat - fit only on training data
scaler.fit(X_train)
X_train = scaler.transform(X_train)
# apply same transformation to test data
X_test = scaler.transform(X_test)
```

Build a Neural Network

```python


clf = MLPClassifier(solver='adam', alpha=1e-5,
                    hidden_layer_sizes=(9, 2), random_state=1)
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python

# Perform the element-wise comparison
# (array1 == array2) returns a Series of True/False values
# .mean() treats True as 1 and False as 0, giving the proportion of True values
match_percentage = (clf.predict(X_test) == y_test).mean() * 100

print(f"Array 1: {list(clf.predict(X_test))}")
print(f"Array 2: {list(y_test)}")
print(f"Percentage of elements that are the same: {match_percentage:.2f}%")
```
